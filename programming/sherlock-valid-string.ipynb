```python
def isValid(s) :
    alphabets = [0] * 26
    for ch in s : 
        index = ord(ch) - 97
        val = alphabets[index]    
        alphabets[index] = alphabets[index]+1

    alphabets_dict = {}
    for i in range(26) :
        if alphabets[i] == 0 :
            continue
        char = chr(i+97)
        alphabets_dict[char] = alphabets[i]
        
    li = [ l for l in alphabets if l!=0 ]
    li = list(set(li))

    result = False
    if len(li) > 2:
        result = False
    elif len(li) == 1 and li[0] == 1:
        result = True
    else :
        count1 = 0
        count2 = 0 
        for ch in alphabets_dict :
            if alphabets_dict[ch] == li[0] :
                count1 = count1+1
            if alphabets_dict[ch] == li[1] :
                count2 = count2+1

        if count1 == 1 :
            result = True
        elif count2 == 1 and li[0]+1 == li[1] :
            result = True
    return result
```
